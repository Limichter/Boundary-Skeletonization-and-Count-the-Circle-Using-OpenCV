```python
#import library
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#membaca file circle.png
img = cv2.imread('circles.png', 0)

#mendapatkan ukuran gambar
m,n = img.shape

#menampilkan gambar input
plt.imshow(img, cmap="gray")
```

# Boundary

```python
#menentukan structuring element
k = 7
kernel = np.ones((k,k),np.uint8)
constant = (k-1)//2

#menampilkan kernel
kernel
```

Kernel menentukan jumlah pixel tetangga yang akan masuk dalam perhitungan. Elemen struktur yang lebih besar memiliki efek yang lebih jelas.
Sedangkan hasil erosi dengan elemen struktur yang besar sama dengan hasil yang diperoleh dari erosi berulang menggunakan elemen struktur yang lebih kecil dengan bentuk yang sama.

```python
#membuat array yang seukuran untuk menampung pixel image
#yang sudah dimodifikasi dengan structuring element
imgErode= np.zeros((m,n), dtype=np.uint8)
```

Meggunakan fungsi morfologi implementasi manual

```python
#melakukan erosion dengan implementasi manual (tanpa fungsi opencv)
for i in range(constant,m-constant):
    for j in range(constant, n-constant):
        temp = img[i-constant:i+constant+1, j-constant:j+constant+1]
        product = temp*kernel
        #erode mengecilkan gambar sehingga 
        #value yang diambil merupakan value terkecil dari pixel
        imgErode[i,j]=np.min(product) 
```

```python
#menampilkan gambar fungsi manual
plt.imshow(imgErode, cmap="gray")
```

Menggunakan morfolgi fungsi opencv

```python
erosion = cv2.erode(img,kernel,iterations = 1)
plt.imshow(erosion, cmap="gray")
```

Dari erosi yang telah dilakukan, didapatkan hasil bulatan yang lebih kecil. erosi menghilakan detail kecil dari gambar biner dan juga mengurangi ukuran wilayah yang diinginkan.
Hal ini dapat dengan mengambil value yang paling kecil. Dengan mengurangkan gambar asli kepada gambar erosi dan didapatkanlah boundary (batas gambar)

```python
#boundary subtraction dapat dicapai dengan 
#melakukan substract image original dengan image hasil erosi
newImg = img - imgErode
```

```python
#menampilkan hasil boundary
plt.imshow(newImg, cmap="gray")
```

```python
#menjadikan gambar ke jpg
cv2.imwrite("hasilBoundary.jpg",newImg)
```

# Skeletonization

Skeleton adalah suatu dari struktur bentuk
Skeletoniztion adalah suatu proses mempertahankan struktur bentuk tetapi semua pixel yang berlebih harus dibuang

Untuk melakukan skeletonization, pertama-pertama harus melakukan operasi open pada gambar. Open merupakan erosi yang diikuti oleh dilatasi. Open dapat membuka celah antar objek yang dihubungkan oleh jembatan tipis piksel. Opening biasanya digunakan untuk menghilangkan objek-objek kecil dan kurus serta dapat membuat tepi citra lebih smooth (untuk citra berukuran besar).
Setelah gambar di open, gambar asli akan di substract atau dikurangi dengan gambar open.

```python
#membaca file circle.png
img2 = cv2.imread('circles.png', 0)

#mendapatkan ukuran gambar
m,n = img2.shape
```

Skeletonization menggunakan morfologi dengan implementasi manual

```python
#membuat array sebesar ukuran gambar
#untuk menampung pixel image
size = np.size(img2)
imgSkel= np.zeros((m,n), dtype=np.uint8)

#structuring element berbentuk persegi dengan dimensi 3x3
element = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
constant = (3-1)//2
eroded = np.zeros((m,n), dtype=np.uint8)
temp = np.zeros((m,n), dtype=np.uint8)

while True :
    #erosi gambar
    for i in range(constant,m-constant):
        for j in range(constant, n-constant):
            temps = img2[i-constant:i+constant+1, j-constant:j+constant+1]
            product = temps*element
            eroded[i,j]=np.min(product)
    #dilatasi gambar dari hasil erosi
    for i in range(constant,m-constant):
        for j in range(constant, n-constant):
            temps = eroded[i-constant:i+constant+1, j-constant:j+constant+1]
            product = temps*element
            temp[i,j]=np.max(product)
    #mengurangi gambar asli dengan gambar dilatasi
    temp = img2 - temp

    imgSkel = cv2.bitwise_or(imgSkel, temp)
    img2 = eroded.copy()

    #Jika tidak ada pixel putih yang tersisa(pixel putih nilainya 255)
    #maka gambar sudah berhasil dierosi
    #keluar dari loop
    zeros = (np.size(img2)) - cv2.countNonZero(img2)
    if zeros==size:
        break
```

```python
plt.imshow(imgSkel, cmap="gray")
```

Skeletonization menggunakan morfologi dari fungsi opencv

```python
img3 = cv2.imread('circles.png', 0)
m,n = img3.shape
size = np.size(img3)
imgSkel= np.zeros((m,n), dtype=np.uint8)

#structuring element berbentuk persegi dengan dimensi 3x3
element = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))

while True :
    #erosi gambar
    eroded = cv2.erode(img3, element)
    #dilatasi gambar dari hasil erosi
    temp = cv2.dilate(eroded, element)
    #mengurangi gambar asli dengan gambar dilatasi
    temp = cv2.subtract(img3, temp)

    imgSkel = cv2.bitwise_or(imgSkel, temp)
    img3 = eroded.copy()

    #Jika tidak ada pixel putih yang tersisa(pixel putih nilainya 255)
    #maka gambar sudah berhasil dierosi
    #keluar dari loop
    zeros = (np.size(img3)) - cv2.countNonZero(img3)
    if zeros==size:
        break
        
plt.imshow(imgSkel, cmap="gray")
```

```python
#menjadikan gambar ke jpg
cv2.imwrite("hasilSkel.jpg",imgSkel)
```

# Mehintung Jumlah Circle
Menggunakan morfologi dan conected component

```python
img = cv2.imread('pcb.jpg')
plt.imshow(img)
```

```python
#mengubah gambar menjadi grayscale dan di threshold
imgGry = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, imgbin = cv2.threshold(imgGry, 240 , 255, cv2.CHAIN_APPROX_NONE)
m,n = imgbin.shape
```

```python
#menentukan structuring element
k = 15
kernel = np.ones((k,k),np.uint8)
constant = (k-1)//2
#diambil 15 agar didapatkan kernel yang tepat

#membuat array yang seukuran untuk menampung pixel image
#yang sudah dimodifikasi dengan structuring element
dilation = np.zeros((m,n), dtype=np.uint8)
```

```python
#fungsi dilatasi dengan implementasi manual
for i in range(constant,m-constant):
        for j in range(constant, n-constant):
            temp = imgbin[i-constant:i+constant+1, j-constant:j+constant+1]
            product = temp*kernel
            dilation[i,j]=np.min(product)
#disini diambil np.min dikarenakan objek telah di threshold sebelumnya
```

```python
#menampilkan hasil dilatasi
plt.imshow(dilation)
```

```python
#menampilkan hasil akhir dilatasi
imgs = cv2.cvtColor(dilation, cv2.COLOR_GRAY2BGR)
_, imgbins = cv2.threshold(imgs, 200 , 256, cv2.CHAIN_APPROX_NONE)
plt.imshow(imgbins)
```

```python
#mencari connectedComponent pada hasil dilatasi
nb_components, output, stats, centroids  = cv2.connectedComponentsWithStats(dilation, connectivity = 8)
```

```python
#mencari pusat dari lingkaran
circlecenter = []

sizes = stats[:, -1]
for i in range(2, nb_components):
        if sizes[i] < 250 and sizes[i] > 200:
            print(stats[i])
            circlecenter.append(centroids[i])
print("Lingkarang yang terdeteksi : ", len(circlecenter))
```

```python
#menampilkan lingkaran yang telah terdeteksi
dilatedcopy = imgbins.copy()
for x in circlecenter:
    tup = (
        int(x[0]),
        int(x[1])
    )
    cv2.circle(dilatedcopy, tup , 10, (0,0,255), -1)
plt.imshow(dilatedcopy)
```

```python
#menampilkan hasil lingkaran dengan gambar awal
out = img.copy()
for x in circlecenter:
    tup = (
        int(x[0]),
        int(x[1])
    )
    cv2.circle(out, tup , 10, (0,0,255), -1)
plt.imshow(out)
```

```python
#menjadikan gambar ke jpg
cv2.imwrite("hasilOut.jpg",out)
```
